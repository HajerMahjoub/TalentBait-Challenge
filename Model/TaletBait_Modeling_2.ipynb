# LinearSVC ¶

```

SVC_pipeline = Pipeline([
                ('tfidf', TfidfVectorizer(stop_words=stop_words)),
                ('clf', OneVsRestClassifier(LinearSVC(), n_jobs=1)),
            ])
```

```
for category in categories:
    print('... Processing {}'.format("label_"+category))
    # train the model using X_dtm & y
    SVC_pipeline.fit(X_train, train["label_"+category])
    # compute the testing accuracy
    prediction = SVC_pipeline.predict(X_test)
    print('Test accuracy is {}'.format(accuracy_score(test["label_"+category], prediction)))
    print('Confusion Matrix is {}'.format(confusion_matrix(test["label_"+category], prediction)))
    print('classification report is {}'.format(classification_report(test["label_"+category], prediction)))
```
